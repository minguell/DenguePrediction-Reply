```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
base_path = os.getcwd()
months = list(range(1, 6 * 12 + 1))
df_input = pd.read_csv(base_path + "/data/df_model_prediction_POA_REPLICATION_YEAR.csv")
```

```python
df_bairro = df_input[df_input['nome_distrito'] == 'EIXO BALTAZAR']
dengue_diagnosis = list(df_bairro["dengue_diagnosis"])
dengue_diagnosis_previsto = list(df_bairro["dengue_diagnosis_previsto"])

plt.plot(months, dengue_diagnosis, label = 'Real')
plt.plot(months, dengue_diagnosis_previsto, label = 'Predicted')
plt.legend()
plt.title("EIXO BALTAZAR")
plt.show()
```
