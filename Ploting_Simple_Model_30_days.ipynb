```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
```

```python
base_path = os.getcwd()
months = list(range(1,7*12 + 1)) #4 anos + 10 meses
df_input = pd.read_csv(base_path+"/data/result_prediction_simple.csv")
```

```python
df_bairro = df_input[df_input['nome_distrito'] == 'CRISTAL']
dengue_diagnosis = list(df_bairro["dengue_diagnosis"])
dengue_diagnosis_previsto = list(df_bairro["dengue_diagnosis_previsto"])
plt.plot(months,dengue_diagnosis,label = 'Real')
plt.plot(months,dengue_diagnosis_previsto,label = 'Predicted')
plt.legend()
locs, labels = plt.xticks()
plt.ylabel("Dengue casese")
plt.xticks(np.arange(0, 84,step = 12))  # Set label locations.
plt.xticks([0, 12, 24,36,48,60,72, 84], ['2017', '2018', '2019','2020', '2021', '2022','2023', '2024'],
       rotation = 20)
#plt.xticks([0, 1, 2,3,4,5], ['January', 'February', 'March','January', 'February', 'March'],
#       rotation=20)  # Set text labels and properties.
plt.title("Cristal")
plt.show()
df_print = df_bairro[df_bairro["ano"] == 2023]
df_print = df_print[df_print["mes"] == 12]
print(df_print)
```
